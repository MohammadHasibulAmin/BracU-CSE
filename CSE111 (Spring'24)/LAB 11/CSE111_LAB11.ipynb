```python
#Task 2
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        self.area = 0

    def calculate_area(self):
        return self.area
    
    def print_details(self):
        print("--------- Printing details ----------")
        print(f'Co-ordinate: ({self.x},{self.y})')
        print(f'Area: {self.area}')

class Circle(Point):
    def __init__(self, radius, x = 0, y= 0):
        super().__init__(x, y)
        self.radius = radius
        self.area = 3.1416 * (radius ** 2)
    
    def print_details(self):
        super().print_details()
        print(f"Radius: {self.radius}")

class Sphere(Circle):
    def __init__(self, radius, x = 0, y = 0):
        super().__init__(radius, x, y)
        self.area = 4 * 3.1416 * (radius ** 2)

print("--------------1---------------")
p1 = Point(2,3)
print(f'Area of p1: {p1.calculate_area()}')
print("--------------2---------------")
p1.print_details()
print("--------------3---------------")
p2 = Point()
p2.print_details()
print("--------------4---------------")
c1 = Circle(4,0,3)
print(f'Area of c1: {c1.calculate_area()}')
print("--------------5---------------")
c1.print_details()
print("--------------6---------------")
c2 = Circle(7)
print(f'Area of c2: {c2.calculate_area()}')
print("--------------7---------------")
sph1 = Sphere(3,0,2)
print(f'Area of sph1: {sph1.calculate_area()}')
print("--------------8---------------")
sph1.print_details()
print("--------------9---------------")
sph2 = Sphere(6)
print(f'Area of sph2: {sph2.calculate_area()}')
```

```python
#Task 3
class Account:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
        self.account_type = "General"
        self.maturity = 0

    def print_details(self):
        print("------ Account details ------")
        print(f"Account Type: {self.account_type}, Maturity: {self.maturity} years")
        print(f"Account Number: {self.account_number}, Balance: ${self.balance:.2f}")

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount:.2f}. New Balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
        else:
            print("Insufficient funds.")

    def year_passed(self, year):
        self.maturity += year
        print(f"Maturity of the account: {self.maturity} years")

class SavingsAccount(Account):
    def __init__(self, a_type, account_number, balance, int_rate, min_limit):
        super().__init__(account_number, balance)
        self.account_type = a_type
        self.int_rate = int_rate
        self.min_limit = min_limit

    def print_details(self):
        super().print_details()
        print(f"Interest Rate: {self.int_rate}, Minimum Limit: ${self.min_limit}")

    def apply_interest(self):
        self.balance += (self.balance * self.int_rate)
        print(f"Interest applied. New Balance: ${self.balance}")
    
    def withdraw(self, amount):
        if self.balance - amount < self.min_limit:
            print("Insufficient funds.")
        else:
            super().withdraw(amount)     


class FixedDepositAccount(Account):
    def __init__(self, a_type, account_number, balance, mat_age):
        super().__init__(account_number, balance)
        self.account_type = a_type
        self.mat_age = mat_age

    def deposit(self, amount):
        print("You can not deposit in a fixed deposit account.")

    def withdraw(self, amount):
        if self.maturity < self.mat_age:
            print("Can not withdraw, Account is not matured")
        else:
            if self.balance >= amount:
                self.balance -= amount
                print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
            else:
                print("Insufficient funds.")

print("-----------1------------")
account = Account("A203", 2000)
account.print_details()
print("-----------2------------")
account.deposit(400)
account.withdraw(1500)
account.year_passed(2)
print("-----------3------------")
account.print_details()
print("-----------4------------")
savings_account = SavingsAccount("Savings","SA123", 1000, 0.05, 500)
savings_account.print_details()
print("-----------5------------")
savings_account.deposit(400)
print("-----------6------------")
savings_account.withdraw(1000)
print("-----------7------------")
savings_account.withdraw(800)
print("-----------8------------")
savings_account.apply_interest()
print("-----------9------------")
savings_account.print_details()
print("-----------10------------")
fixed_account1= FixedDepositAccount("Fixed Deposit","FDA321", 10000, 5)
fixed_account1.print_details()
print("-----------11------------")
fixed_account1.deposit(400)
print("-----------12------------")
fixed_account1.year_passed(6)
print("-----------13------------")
fixed_account1.withdraw(10000)
print("-----------14------------")
fixed_account1.print_details()
print("-----------15------------")
fixed_account2 = FixedDepositAccount("Fixed Deposit","FDA300", 50000, 7)
fixed_account2.print_details()
print("-----------16------------")
fixed_account2.withdraw(10000)
```

```python
#Task 2
class Exam:
    def __init__(self,marks): 
        self.marks = marks 
        self.time = 60 
         
    def examSyllabus(self): 
        return "Maths , English" 
    def examParts(self): 
        return "Part 1 - Maths\nPart 2 - English\n"

class ScienceExam(Exam):
    def __init__(self, marks, time, *args):
        super().__init__(marks)
        self.time = time
        self.syllabus = []
        self.parts = 2
        for i in args:
            self.parts += 1
            self.syllabus.append(i)
            
    def examSyllabus(self): 
        return f"{super().examSyllabus()} , {' , '.join(self.syllabus)}"
    
    def examParts(self):
        printer = ""
        count = 3

        for i in range(len(self.syllabus)):
            printer += f"Part {count} - {self.syllabus[i]}\n"
            count += 1
            
        return super().examParts() + printer

    def __str__(self):
        return f"Marks: {self.marks} Time: {self.time} minutes Number of Parts: {self.parts}"
        

engineering = ScienceExam(100,90,"Physics","HigherMaths") 
print(engineering) 
print('----------------------------------') 
print(engineering.examSyllabus()) 
print(engineering.examParts()) 
print('==================================')  
architecture =  ScienceExam(100,120,"Physics","HigherMaths","Drawing") 
print(architecture) 
print('----------------------------------') 
print(architecture.examSyllabus()) 
print(architecture.examParts())
```

```python
#Task 3
class PokemonBasic:
  def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
    self.name = name 
    self.hit_point = hp 
    self.weakness = weakness 
    self.type = type 
 
  def get_type(self):
    return 'Main type: ' + self.type 
 
  def get_move(self): 
    return 'Basic move: ' + 'Quick Attack' 
 
  def __str__(self): 
    return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness 
 
class PokemonExtra(PokemonBasic):
  def __init__(self, name, hp, weakness, type = None, s_type = None, extra_moves = None):
    super().__init__(name, hp, weakness, type)
    self.s_type = s_type
    self.extra_moves = extra_moves

    if extra_moves != None:
      #self.extra_moves = list(extra_moves)
      self.extra_moves = ', '.join(extra_moves)

  def get_type(self):
    if self.s_type != None:
      return f"{super().get_type()}, Secondary type: {self.s_type}"
    return super().get_type()
        
  def get_move(self):
    if self.extra_moves != None:
      return f"{super().get_move()} \nOther move: {self.extra_moves}"
    return super().get_move()
  

print('\n------------Basic Info:--------------') 
pk = PokemonBasic() 
print(pk) 
print(pk.get_type()) 
print(pk.get_move()) 
 
print('\n------------Pokemon 1 Info:-------------') 
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire') 
print(charmander) 
print(charmander.get_type()) 
print(charmander.get_move()) 
 
print('\n------------Pokemon 2 Info:-------------') 
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze')) 
print(charizard) 
print(charizard.get_type()) 
print(charizard.get_move())
```

```python
#Task 4
class Cakes:
    order_list = {}
    feedbacks = {}
    def __init__(self, flavor, weight):
        self.flavor = flavor + " Cake"
        self.weight = weight
        self.price = 1200 * (self.weight/1000)

        self.sweetness = "Moderate sugar"
        self.cream = "Whipped Cream"

        self.appender = f"{self.flavor} {str(self.weight)} gm"
        
        if self.appender not in Cakes.order_list.keys():
            Cakes.order_list[self.appender] = 1
        else:
            Cakes.order_list[self.appender] += 1

    def add_customization(self, sweetness, cream):
        self.sweetness = sweetness + " sugar"
        self.cream = cream

    @classmethod
    def give_feedbacks(cls, flavor, rating):
        print("Thanks for your valuable feedback.")
        # Check if the flavor already exists in the show_feedbacks dictionary
        if flavor in cls.feedbacks:
            # If it exists, append the rating to the list
            cls.feedbacks[flavor].append(rating)
        else:
            # If it doesn't exist, create a new list with the rating
            cls.feedbacks[flavor] = [rating]

    @classmethod
    def show_feedbacks(cls):
        print(cls.feedbacks)

    def cake_details(self):
        print(f"Flavor: {self.flavor}, Weight: {self.weight} gm")
        print(f"Sweetness: {self.sweetness} Cream Type: {self.cream}")
        print(f"Price: {self.price} Taka")

class Cheese_Cakes(Cakes):
    #discount = 1
    def __init__(self, flavor, weight, cake_t = "baked"):
        super().__init__(flavor + " Cheese", weight)

        self.discount = 0

        self.price = 1500 * (self.weight/1000) #* Cheese_Cakes.discount

        self.sweetness = "Moderate sugar"
        self.cream = "Cream Cheese"
        self.cake_t = cake_t

    def add_customization(self):
        print("Sorry! No customization available for cheese cakes.")

    @classmethod
    def give_feedbacks(cls, flavor, rating):
        super().give_feedbacks(flavor, rating)
        print("You will get 10% discount for your next purchase")
        #cls.dicount = 0.1
  
    def cake_details(self):
        print(f"Flavor: {self.flavor}, Weight: {self.weight} gm")
        print(f"Sweetness: {self.sweetness} Cream Type: {self.cream}")
        print(f"Cake Type:{self.cake_t}, Price: {self.price} Taka")

order_1=Cakes("Chocolate",500)
order_2=Cakes("Vanilla",800)
print("(1)**********************************")
order_1.cake_details()
print("(1.1)**********************************")
print(Cakes.order_list)
print("(2)**********************************")
order_2.add_customization("Zero","Butter Cream")
order_2.cake_details()
print("(3)**********************************")
Cakes.give_feedbacks("Chocolate Cake","Very Delicious")
Cakes.give_feedbacks("Chocolate Cake","Yummy")
print("(4)**********************************")
Cakes.show_feedbacks()
print("(5)**********************************")
ch_order1=Cheese_Cakes("Red velvet",700)
ch_order1.cake_details()
print("(6)**********************************")
ch_order1.add_customization()
print("(7)**********************************")
ch_order2=Cheese_Cakes("Blue Berry",900,"No Bake")
ch_order2.cake_details()
print("(8)**********************************")
print(Cakes.order_list)
print("(9)**********************************")
Cheese_Cakes.give_feedbacks("Red velvet Cheese Cake","Average")
Cheese_Cakes.give_feedbacks("Blue Berry Cheese Cake","Liked it")
print("(10)**********************************")
Cakes.show_feedbacks()
```
