```python
#Task 1
class Circle:
    def __init__(self, radius):
        self.__radius = radius
        self.__area = 0
    
    def getRadius(self):
        return self.__radius
    
    def area(self):
        self.__area = 3.1416 * (self.__radius ** 2)
        return self.__area

c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" , c1.area())


c2 = Circle(5)
print("Second circle radius:" , c2.getRadius())
print("Second circle area:" , c2.area())
```

```python
#Task 2
class Triangle:
  def __init__(self, base, height):
    self.__base = base
    self.__height = height
    self.__area = 0

  def getBase(self):
    return self.__base
  def getHeight(self):
    return self.__height
  def setBase(self, base):
    self.__base = base
  def setHeight(self, height):
    self.__height = height

  def area(self):
    self.__area = 0.5 * self.__base * self.__height
    return self.__area

#Driver Code
t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())


t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
#Task 3
class Team:
  def __init__(self, team_n = None):
    self.__team = team_n
    self.__player_list = []

  def setName(self, name):
    self.__team = name
    return self.__team

  def printDetail(self):
    print(f"Team : {self.__team}")
    print(f"List of Players : {self.__player_list}")

  def addPlayer(self, obj):
    self.__player_list.append(obj.name)

class Player:
  def __init__(self, name):
    self.name = name

#Driver Code
b = Team()
b.setName('Bangladesh')
mashrafi = Player("Mashrafi")
b.addPlayer(mashrafi)
tamim = Player("Tamim")
b.addPlayer(tamim)
b.printDetail()
a = Team("Australia")
ponting = Player("Ponting")
a.addPlayer(ponting)
lee = Player("Lee")
a.addPlayer(lee)
a.printDetail()
```

```python
#Task 4
class Spaceship:
  def __init__(self, name, capacity):
    self.name = name
    self.capacity = capacity
    self.cargo_list = []
    self.total_weight = 0

  def load_cargo(self, obj):
      if self.total_weight + obj.getweight() < self.capacity:
        self.cargo_list.append(obj.grab_item())
        self.total_weight += obj.getweight()
      else:
        print(f"Warning : unable to load {obj.grab_item()} inside {self.name}. \nExceeds capacity by {self.total_weight + obj.getweight() - self.capacity}")

  def display_details(self):
      print(f"Spaceship Name : {self.name}")
      print(f"Capacity : {self.capacity}")
      print(f"Current Cargo Weight : {self.total_weight}")
      print(f"Cargo : {self.cargo_list}")

class Cargo:
  def __init__(self, item, weight):
    self.__item = item
    self.__weight = weight

  def grab_item(self):
    return self.__item

  def getweight(self):
    return self.__weight

# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#Task 5
class BracuStudent:
    def __init__(self, name, home):
        self.name = name
        self.home = home
        self.bus_pass = False
    
    def get_pass(self):
        self.bus_pass = True

    def show_details(self):
        print(f"Student Name: {self.name}")
        print(f"Lives in {self.home}")
        print(f"Have a Bus Pass? {self.bus_pass}")

class BracuBus:
    def __init__(self, bus_loc, capacity = 2):
        self.bus_loc = bus_loc
        self.capacity = capacity
        self.pass_count = 0
        self.pass_list = []
    
    def board(self, *args):
        if len(args) == 0:
            print("No passengers!")
        else:
            for i in args:
                if self.pass_count < self.capacity:
                    if i.bus_pass == True:
                        if i.home == self.bus_loc:
                            print(f"{i.name} boarded the bus")
                            self.pass_list.append(i.name)
                            self.pass_count += 1
                        else:
                            print("You got on the wrong bus!")
                    else:
                        print("You don't have a bus pass!")
                else:
                    print("Bus is full!")

    
    def show_details(self):
        print(f"Bus Route : {self.bus_loc}")
        print(f"Passengers Count : {self.pass_count} (Max : {self.capacity})")
        print(f"Passengers On Board : {self.pass_list}")

st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
#Task 6
class Library:
    def __init__(self, location, books):
        self.location = location
        self.borrowers = {}
        self.books = books

    def details(self):
        print(f"{self.location} Library details")
        print("Borrower details:")
        print(self.borrowers)
        print("Books availability:")
        print(self.books)

class Reader:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = {}

    def borrow(self, library, *book_types):
        for book_type in book_types:
            if book_type in library.books and library.books[book_type] > 0:
                if self.name in library.borrowers:
                    if library.borrowers[self.name] + 1 <= 5:
                        library.borrowers[self.name] += 1
                        library.books[book_type] -= 1
                        if book_type in self.borrowed_books:
                            self.borrowed_books[book_type] += 1
                        else:
                            self.borrowed_books[book_type] = 1
                        print(f"{book_type} book is borrowed successfully.")
                    else:
                        print("You cannot borrow more than 5 books.")
                else:
                    library.borrowers[self.name] = 1
                    library.books[book_type] -= 1
                    self.borrowed_books[book_type] = 1
                    print(f"{book_type} book is borrowed successfully.")
            else:
                print(f"{book_type} books are not available at the moment.")

    def readerInfo(self, book_type= None):
        if book_type:
            if book_type in self.borrowed_books:
                print(f"{self.name}, you have {self.borrowed_books[book_type]} {book_type} book(s) with you.")
            else:
                print(f"{self.name}, you have 0 {book_type} book(s) with you.")
        else:
            if self.borrowed_books:
                print(f"{self.name}, you have {sum(self.borrowed_books.values())} book(s) with you.")
                for book_type, quantity in self.borrowed_books.items():
                    print(f"Books on {book_type}: {quantity}")
            else:
                print(f"{self.name}, you have 0 books with you.")

L1 = Library('Dhaka', {'Arts': 15, 'Fiction': 135, 'Politics': 2, 'Science': 11, 'Poetry': 15})
L1.details()
print("1----------------------")
r1 = Reader('Aladdin')
r1.borrow(L1, 'Arts', 'Fiction', 'Fiction', 'Politics')
print("2----------------------")
r1.borrow(L1, 'Politics', 'Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2 = Reader('Jasmine')
r2.borrow(L1, 'Politics', 'Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
