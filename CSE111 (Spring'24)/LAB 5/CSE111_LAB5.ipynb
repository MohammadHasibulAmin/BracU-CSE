```python
#Task 1
class Student:
  def __init__(self,name,id,department = "CSE"):
    self.name = name
    self.id = id
    self.dept = department

  def dailyEffort(self,hours):
    self.hours = hours
    if self.hours <= 2:
      self.effort = "Should give more effort!"
    elif self.hours <= 4:
      self.effort = "Keep up the good work!"
    else:
      self.effort = "Excellent! Now motivate others."

  def printDetails(self):
    print(f"Name : {self.name}")
    print(f"ID : {self.id}")
    print(f"Daily Effort : {self.hours} hour(s)")
    print(f"Suggestion : {self.effort}")

#Driver Code
harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
#Task 2
class Farmer:
  def __init__(self,name = None):
    self.name = name
    self.crops = []
    self.fishes = []
    self.crop_count = 0
    self.fish_count = 0
    if type(self.name) == str:
      print(f"Welcome to your farm,{self.name}!")
    elif type (self.name) == int:
      print(f"Welcome to your farm. Your farm ID is {self.name}!")
    else:
      print("Welcome to your farm")

  def addCrops(self, *crops):
    crop_count = 0
    for i in crops:
      crop_count += 1
      self.crops.append(i)
    if crop_count == 0:
      print("No crop(s) added")
    else:
      print(f"{crop_count} crop(s) added")
    self.crop_count = crop_count

  def addFishes(self, *fishes):
    fish_count = 0
    for i in fishes:
      fish_count += 1
      self.fishes.append(i)
    if fish_count == 0:
      print("No fish(s) added")
    else:
      print(f"{fish_count} fish(s) added")
    self.fish_count = fish_count

  def showGoods(self):
    if self.crops != []:
      print(f"You have {self.crop_count} crop(s) :")
      for i in range(len(self.crops)):
        if i == len(self.crops) - 1:
          print(self.crops[i], end = "")
        else:
          print(self.crops[i], end = ",")
    else:
        print("You don't have any crop(s)")
    print()

    if self.fishes != []:
      print(f"You have {self.fish_count} fish(s) :")
      for i in range(len(self.fishes)):
        if i == len(self.fishes) - 1:
          print(self.fishes[i], end = "")
        else:
          print(self.fishes[i], end = ",")
    else:
      print("You don't have any fish(s)")
    print()

f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
#Task4

class Sphere:
  def __init__(self, id, radius = 1, color = 'White'):
    self.id = id
    self.radius = radius
    self.color = color
    self.volume = (4/3) *(3.1416)* (radius**3)
  def printDetails(self):
    print(f'Sphere ID: {self.id}')
    print(f'Color: {self.color}')
    print(f'Volume: {self.volume}')
  def merge_sphere(self, *spheres):
    #print(spheres)
    print('Spheres are being merged')
    for i in spheres:
      #print(i.volume)
      self.volume += i.volume
      if self.color != i.color:
        self.color = 'Mixed Color'

sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1,sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()
```

```python
#Task 1
class Student:
    def __init__(self, name, cgpa, creds = 9, dept = "CSE"):
        self.name = name
        self.cgpa = cgpa
        self.creds = creds
        self.dept = dept
    
    def checkScholarshipEligibility(self):
        self.schol_status = ""
        if self.creds > 10 and self.cgpa >= 3.5:
            if self.cgpa >= 3.7:
                print(f"{self.name} is eligible for Merit based scholarship")
                self.schol_status = "Merit based scholarship"
            elif 3.7 > self.cgpa >= 3.5:
                print(f"{self.name} is eligible for need based scholarship")
                self.schol_status = "Need-based scholarship"
        else:
            print(f"{self.name} is not eligible for scholarship")
            self.schol_status = "No scholarship"

    def showDetails(self):
        print(f"Name : {self.name}")
        print(f"Department : {self.dept}")
        print(f"CGPA : {self.cgpa}")
        print(f"Number of credits : {self.creds}")
        print(f"Scholarship Status : {self.schol_status}")
                

print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
#Task 3
class Department:
    def __init__(self, dept="ChE Department", sections=5):
        self.department = dept
        self.sections = sections
        self.total_students = 0
        self.avg_students = 0

        print(f"The {self.department} has {self.sections} sections.")

    def add_students(self, *args):
        if len(args) != self.sections:
            print(f"The {self.department} doesn't have {len(args)} sections.")
        else:
            for i in args:
                self.total_students += i
            self.avg_students = self.total_students / self.sections
            self.avg_students = round(self.avg_students, 2)
            print(f"The {self.department} has an average of {self.avg_students} students in each section.")

    def merge_Department(self, *deptObj):
        for i in deptObj:
            print(f"{i.department} is merged to {self.department}")
            self.total_students += i.avg_students * i.sections

        self.avg_students = self.total_students / self.sections
        return f"Now the {self.department} has an average of {self.avg_students:.1f} students in each section."

    
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22,
32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,
51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
#Task 5
class Author:
    def __init__(self, name=None):
        self.name = name
        self.books = []

    def setName(self, namesetter):
        if self.name == None:
            self.name = namesetter

    def addBook(self, b_name, genre):
        if self.name == None:
            print("A book cannot be added without an author name")
        else:
            self.books.append((b_name, genre))

    def printDetail(self):
        print(f"Number of Book(s): {len(self.books)}")
        print(f"Author Name: {self.name}")
        for book in self.books:
            print(f"{book[1]}: {book[0]}")
        
a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
