```python
#Task 1
class Customer:
    def __init__(self):
        print("Welcome to ABC Memorial Park")
        self.tc = 0
        self.price = 0
    def buyTicket(self, name, age):
        if self.tc < 3:
            self.tc += 1
            print(f"Successfully purchased a ticket for {name}!")
            if age > 10:
                self.price += 100
            else:
                self.price += 50
        else:
            print("you can't buy more than 3 tickets")
    def showDetails(self):
        print(f"Amount of tickets : {self.tc}")
        print(f"Total price : {self.price} Taka")

print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
#Task 2
class Panda:
    def __init__(self,name,gender,age):
        self.name = name
        self.gender = gender
        self.age = age
    def sleep(self, hours):
        if 3<hours<=5:
            return (f"{self.name} sleeps {hours} hours daily and should have Mixed Veggies")
        elif 6<hours<=8:
            return (f"{self.name} sleeps {hours} hours daily and should have Eggplant and Tofu")
        elif 9<hours<=11:
            return (f"{self.name} sleeps {hours} hours daily and should have Broccoli Chicken")
        else:
            return (f"{self.name}'s duration is {hours} thus should have only bamboo leaves")
        
panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female",3)
panda3 = Panda("Ming Ming", "Female",8)

print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))

print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))

print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep(13))
```

```python
#Task 3
class GreenPhone:
    def __init__(self, name, version, cameras):
        self.company = "GreenPhone"
        self.model_name = name
        self.version = version
        self.cameras = cameras
        self.update_count = 0

        if "A" in name:
            self.max_update = 2
        elif "M" in name:
            self.max_update = 3
        elif "U" in name:
            self.max_update = 4

    def updatePhone(self):
        if self.update_count < self.max_update:
            self.update_count += 1
            self.version += 1
            print(f"Your phone {self.company} {self.model_name} is upgraded to Android Version: {self.version}.")
        else:
            print(f"Your phone {self.company} {self.model_name} is already up to date.")

    def showSpecification(self):
        print(f"Phone Company: {self.company}")
        print(f"Model Name: {self.model_name}")
        print(f"Android Version: {self.version}")
        print(f"Number of Cameras: {self.cameras}")

print('1=======================')
p1 = GreenPhone('A1', 12, 3)
p2 = GreenPhone('M11', 12, 4)
p3 = GreenPhone('U20', 12, 5)
p1.showSpecification()
print('2=======================')
p2.showSpecification()
print('3=======================')
p1.updatePhone()
print('4=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('5=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('6=======================')
p2.updatePhone()
p3.updatePhone()
print('7=======================')
p1.showSpecification()
p3.showSpecification()
```

```python
class StudentDatabase:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = {}  # Dictionary to store grades for each semester

    def calculateGPA(self, course_grades, semester):
        total_credit_points = 0
        courses_tuple = tuple()
        
        # Extract course names and grades from the input list
        for course_info in course_grades:
            course, grade = course_info.split(": ")
            courses_tuple += (course,)
            total_credit_points += float(grade) * 3
        
        # Calculate GPA for the semester
        semester_gpa = round(total_credit_points / (len(course_grades) * 3), 2)
        
        # Store the calculated GPA for the semester
        self.grades[semester] = {courses_tuple: semester_gpa}

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.student_id}")

        # Iterate through each semester in the grades dictionary
        for semester, courses_info in self.grades.items():
            print(f"Courses taken in {semester}:")
            
            # Extract course names and GPA for the semester
            for courses_tuple, semester_gpa in courses_info.items():
                for course_name in courses_tuple:
                    print(course_name)
                
                # Print the calculated GPA for the semester
                print(f"CGPA: {semester_gpa}")

# Example Usage
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()

s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```
