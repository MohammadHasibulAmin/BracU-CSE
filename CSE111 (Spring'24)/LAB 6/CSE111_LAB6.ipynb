```python
#Task 1
class Teacher:
  def __init__(self, name , dept):
    self.name = name
    self.dept = dept
    self.courses_list = []

  def addCourse(self, obj):
      self.courses_list.append(obj.course)

  def printDetail(self):
    print(f"Name : {self.name}")
    print(f"Department : {self.dept}")
    print("List of courses : ")
    print("="*40)
    for i in (self.courses_list):
      print(i)
    print("="*40)
    print("="*40)

class Course:
  def __init__(self, course):
    self.course = course

#Driver Code
t1 = Teacher("Saad Abdullah", "CSE")
t2 = Teacher("Mumit Khan", "CSE")
t3 = Teacher("Sadia Kazi", "CSE")
c1 = Course("CSE 110 Programming Language I")
c2 = Course("CSE 111 Programming Language-II")
c3 = Course("CSE 220 Data Structures")
c4 = Course("CSE 221 Algorithms")
c5 = Course("CSE 230 Discrete Mathematics")
c6 = Course("CSE 310 Object Oriented Programming")
c7 = Course("CSE 320 Data Communications")
c8 = Course("CSE 340 Computer Architecture")
t1.addCourse(c1)
t1.addCourse(c2)
t2.addCourse(c3)
t2.addCourse(c4)
t2.addCourse(c5)
t3.addCourse(c6)
t3.addCourse(c7)
t3.addCourse(c8)
t1.printDetail()
t2.printDetail()
t3.printDetail()
```

```python
#Task 2
class Student:
  def __init__(self, name , id, cgpa):
    self.name = name
    self.id = id
    self.cgpa = cgpa

  def setId(self, id):
    self.id = id

class Department:
  def __init__(self, dept):
    self.dept = dept
    self.NAME = []
    self.ID = []
    self.CGPA = []
    self.count = 0

  def findStudent(self, finder):
    if finder in self.ID:
      print("Student info :")
      print(f"Student Name : {self.NAME[self.ID.index(finder)]}")
      print(f"ID : {finder}")
      print(f"CGPA : {self.CGPA[self.ID.index(finder)]}")
      print(f"Welcome to CSE department, {self.NAME[self.ID.index(finder)]}")
    else:
      print("Student with this ID doesn't exist, Please give a valid ID")

  def addStudent(self, *args):
    for i in args:
      if i.id not in self.ID:
        self.ID.append(i.id)
        self.NAME.append(i.name)
        self.CGPA.append(i.cgpa)
        self.count += 1
        print(f"Welcome to CSE department, {i.name}")
      else:
        print("Student with the same ID already exists, Please try with another ID")

  def details(self):
    print("Department Name: CSE")
    print(f"Number of students : {self.count}")
    for i in range(len(self.ID)):
      print(f"Student Name : {self.NAME[i]}, ID : {self.ID[i]}, cgpa : {self.CGPA[i]}")

#Driver Code
s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
#Task 3
class Vaccine:
  def __init__(self,name,place,validity):
    self.name=name
    self.place=place
    self.validity=validity

class Person:
  def __init__(self,name,age,job="General Citizen"):
    self.name=name
    self.age=age
    self.job=job
    self.vac=None
    self.d1=""
    self.d2="Please come after 60 days"

  def pushVaccine(self,location,d="1st Dose"):
    if (self.job=="Student" or self.age>=25):
      if self.vac==None or self.vac==location:
        self.vac=location
        if d=="1st Dose":
          self.d1="Given"
        elif d=="2nd Dose":
          self.d2="Given"
        print(f"{d} done for Bob")
      else:
        print(f"Sorry {self.name}, you can't take 2 different vaccines")
    else:
      print(f"Sorry {self.name}, Minimum age for taking vaccines is 25 years now.")

  def showDetail(self):
    print(f"Name: {self.name} Age: {self.age} Type: {self.job}")
    print(f"Vaccine name: {self.vac.name}")
    print(f"1st dose: {self.d1}")
    print(f"2nd dose: {self.d2}")


astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
