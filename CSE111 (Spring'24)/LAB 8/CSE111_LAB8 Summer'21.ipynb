```python
#Task 1
class Student:
    id_count = 0
    def __init__(self, name, dept, age, cg):
        self.name = name
        self.dept = dept
        self.age = age
        self.cg = cg
        Student.id_count += 1
    
    def get_details(self):
        print(f"ID: {Student.id_count}")
        print(f"name: {self.name}")
        print(f"Department: {self.dept}")
        print(f"Age: {self.age}")
        print(f"CGPA: {self.cg}")
    
    @classmethod
    def from_String(cls, str):
        name, dept, age, cg = str.split("-")
        return cls(name, dept, age, cg)
        

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
```

```python
#Task 2
class Assassin:
    count = 0
    def __init__(self, name, rate):
        self.name = name
        self.rate = rate
        Assassin.count += 1
    
    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"Success rate: {self.rate}%")
        print(f"Total number of Assassin: {Assassin.count}")

    @classmethod
    def failureRate(cls, name, f_rate):
        s_rate = 100 - f_rate
        obj = cls(name, s_rate)
        return obj
    
    @classmethod
    def failurePercentage(cls, name, f_perc):
        s_perc = 100 - f_perc
        return cls(name, s_perc)
        
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task 5
class Employee:
    def __init__(self, name, w_p):
        self.name = name
        self.w_p = w_p

    def workingPeriod(self):
        return self.w_p

    @classmethod
    def employeeByJoiningYear(cls, name , j_year):
        w_p = 2021 - j_year
        return cls(name, w_p)

    @staticmethod
    def experienceCheck(w_p, gender):
        if gender == "male" : gender = "he"
        elif gender == "female" : gender = "she"
        
        if w_p < 3: return (f"{gender} is not experienced")
        else: return (f"{gender} is experienced")

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod())
print(employee2.workingPeriod())
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task 7
class Cat:
    Number_of_cats = 0
    def __init__(self, color = "White", status = "sitting"):
        self.color = color
        self.status = status
        Cat.Number_of_cats += 1

    def printCat(self):
        print(f"{self.color} cat is {self.status}")

    def changeColor(self, color):
        self.color = color
        return self.color

    @classmethod
    def no_parameter(cls):
        return cls()

    @classmethod
    def first_parameter(cls, color):
        return cls(color)

    @classmethod
    def second_parameter(cls, status):
        return cls(status)

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 8
class Cylinder:
    radius = 5
    height = 18
    def __init__(self, radius, height):
        print(f"Default radius = {Cylinder.radius} and height = {Cylinder.height}.")

        Cylinder.radius = radius
        Cylinder.height = height

        print(f"Updated: radius = {Cylinder.radius} and height = {Cylinder.height}.")

    @staticmethod
    def area(radius, height):
        area =  (2 * 3.1416 * (radius ** 2)) + (2 * 3.1416 * radius * height)
        print(f"Area: {area}")

    @staticmethod
    def volume(radius, height):
        volume = 3.1416 * (radius ** 2) * height
        print(f"Volume: {volume}")

    @classmethod
    def swap(cls, height, radius):
        return cls(radius, height)
    
    @classmethod
    def changeFormat(cls, str):
        radius, height = str.split("-")
        return cls(float(radius), float(height))


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 10
class SultansDine:
    n_branches = 0
    total_sell = 0
    list1 = []
    def __init__(self, name):
        SultansDine.n_branches += 1
        self.name = name
    
    def sellQuantity(self, quantity):
        self.branch_sell = 0
        
        if quantity < 10 : self.branch_sell = quantity * 300
        elif quantity < 20 : self.branch_sell = quantity * 350
        else: self.branch_sell =  quantity * 400
        
        SultansDine.total_sell += self.branch_sell

    def branchInformation(self):
        print(f"Branch Name: {self.name}, Branch Sell: {self.branch_sell}")

        SultansDine.list1.append(self.name)
        SultansDine.list1.append(self.branch_sell)
        #SultansDine.branch_data.extend([self.location, self.branch_sell])

    @staticmethod
    def details():
        print(f"Total Number of branch(s): {SultansDine.n_branches}")
        print(f"Total Sell : {SultansDine.total_sell}")
        for i in range(0, len(SultansDine.list1), 2):
            print(f"Branch Name: {SultansDine.list1[i]}, Branch Sell: {SultansDine.list1[i+1]}")
            print(f"Branch consists of total sell's : {round((SultansDine.list1[i+1] / SultansDine.total_sell) * 100, 2)}%")

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
