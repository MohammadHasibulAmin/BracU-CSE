```python
#Task6 (Ai)

import random

def playRockPaperScissor(rounds):
    moveslist = ['rock', 'paper', 'scissor']
    man_score, robot_score = 0, 0

    for _ in range(rounds):
        man_moves = input("Your move (rock, paper, or scissor): ").lower()
        robot_moves = random.choice(moveslist)

        if man_moves in moveslist:
            if man_moves == robot_moves:
                continue
            elif ((man_moves == 'rock' and robot_moves == 'scissor') or
                  (man_moves == 'paper' and robot_moves == 'rock') or
                  (man_moves == 'scissor' and robot_moves == 'paper')):
                man_score += 1
            else:
                robot_score += 1
        else:
            print("Invalid move. Please enter rock, paper, or scissor.")

    result_message = "You have won the game" if man_score > robot_score else "Computer has won the game" if man_score < robot_score else "It's a Tie!"
    return f"Your Score: {man_score}\nComputer's Score: {robot_score}\n{result_message}"

result = playRockPaperScissor(rounds=int(input("Enter the number of rounds: ")))
print(result)
```

```python
#Task 4 (Ai)
class StudentDatabase:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = {}  # Dictionary to store grades for each semester

    def calculateGPA(self, course_grades, semester):
        total_credit_points = 0
        courses_tuple = tuple()
        
        # Extract course names and grades from the input list
        for course_info in course_grades:
            course, grade = course_info.split(": ")
            courses_tuple += (course,)
            total_credit_points += float(grade) * 3
        
        # Calculate GPA for the semester
        semester_gpa = round(total_credit_points / (len(course_grades) * 3), 2)
        
        # Store the calculated GPA for the semester
        self.grades[semester] = {courses_tuple: semester_gpa}

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.student_id}")

        # Iterate through each semester in the grades dictionary
        for semester, courses_info in self.grades.items():
            print(f"Courses taken in {semester}:")
            
            # Extract course names and GPA for the semester
            for courses_tuple, semester_gpa in courses_info.items():
                for course_name in courses_tuple:
                    print(course_name)
                
                # Print the calculated GPA for the semester
                print(f"CGPA: {semester_gpa}")

# Example Usage
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()

s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```

```python
# Task 2 (Friend)
class Foodie:
    def __init__(self, name):
        self.name = name
        self.items = []
        self.spent = 0

    def show_orders(self):
        return f"{self.name} has {len(self.items)} item(s) in the cart.\nItems: {self.items}\nTotal spent: {self.spent}."

    def order(self, *items):
        for i in items:
            n = i.split("-")
            print(f"Ordered - {n[0]}, quantity - {n[1]}, price (per unit) - {menu[n[0]]}")
            print("Total price-", menu[n[0]] * int(n[1]))
            self.spent += menu[n[0]] * int(n[1])
            self.items.append(n[0])

    def pay_tips(self, tips=0):
        if tips == 0:
            print("No tips to the waiter.")
        else:
            print(f"Give {tips}/- tips to the waiter.")
            self.spent += tips

menu = {'Chicken Lollipop': 15, 'Beef Nugget': 20, 'Americano': 180, 'Red Velvet': 150, 'Prawn Tempura': 80,
        'Saute Veg': 200}

f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3', 'Beef Nugget-6', 'Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6', 'Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
#Task 3
class Department:
    def __init__(self, dept="ChE Department", sections=5):
        self.department = dept
        self.sections = sections
        self.total_students = 0
        self.avg_students = 0

        print(f"The {self.department} has {self.sections} sections.")

    def add_students(self, *args):
        if len(args) != self.sections:
            print(f"The {self.department} doesn't have {len(args)} sections.")
        else:
            for i in args:
                self.total_students += i
            self.avg_students = self.total_students / self.sections
            self.avg_students = round(self.avg_students, 2)
            print(f"The {self.department} has an average of {self.avg_students} students in each section.")

    def merge_Department(self, *deptObj):
        for i in deptObj:
            print(f"{i.department} is merged to {self.department}")
            self.total_students += i.avg_students * i.sections

        self.avg_students = self.total_students / self.sections
        return f"Now the {self.department} has an average of {self.avg_students:.1f} students in each section."

    
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22,
32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,
51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
#Task 4 (Rafid)
class Shopidify:
    def __init__(self, name = "Guest"):
        self.name = name
        self.itemDict = dict()
        self.transactionCount = 0

        if self.name == "Guest":
            print("Welcome to Shopidify")

        else:
            print(f"Welcome {self.name} to Shopidify")

    def add_to_cart(self, itemName, itemNum = 1):
        if type(itemName) == str:
            self.itemDict[itemName] = itemNum

        elif type(itemName) == list:
            for i in range(0, len(itemName), 2):
                if (i+1) <= len(itemName)-1:
                    self.itemDict[itemName[i]] = itemName[i+1]

                else:
                    break

    def display_cart(self):
        print(f"Items in the cart for {self.name}:")
        for i in self.itemDict.keys():
            print(f"- {i}: {self.itemDict[i]}x")

    def checkout(self):
        print(f"Checkout completed for {self.name}")
        self.transactionCount += 1

    def display_history(self):
        print(f"Purchase history for {self.name}")
        print(f"Transaction: {self.transactionCount}")
        for i in self.itemDict.keys():
            print(f"- {i}: {self.itemDict[i]}x")


#driver code
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
from abc import ABC, abstractmethod
class food(ABC):        #abstract class
    @abstractmethod
    def taste(self):    #declaring abstract method
        pass            #writing anything inside the code doesn't matter
                        #as we will override it later anyway
class pizza(food):      #concrete class
    def size(self):
        print("12-inch pizza")
    def taste(self):    #overriding abstract method
        print("tastes spicy")
p = pizza()
p.size()
p.taste()
```

```python
#Task 6
class Library:
    def __init__(self, location, books):
        self.location = location
        self.borrowers = {}
        self.books = books

    def details(self):
        print(f"{self.location} Library details")
        print("Borrower details:")
        print(self.borrowers)
        print("Books availability:")
        print(self.books)

class Reader:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = {}

    def borrow(self, library, *book_types):
        for book_type in book_types:
            if book_type in library.books and library.books[book_type] > 0:
                if self.name in library.borrowers:
                    if library.borrowers[self.name] + 1 <= 5:
                        library.borrowers[self.name] += 1
                        library.books[book_type] -= 1
                        if book_type in self.borrowed_books:
                            self.borrowed_books[book_type] += 1
                        else:
                            self.borrowed_books[book_type] = 1
                        print(f"{book_type} book is borrowed successfully.")
                    else:
                        print("You cannot borrow more than 5 books.")
                else:
                    library.borrowers[self.name] = 1
                    library.books[book_type] -= 1
                    self.borrowed_books[book_type] = 1
                    print(f"{book_type} book is borrowed successfully.")
            else:
                print(f"{book_type} books are not available at the moment.")

    def readerInfo(self, book_type= None):
        if book_type:
            if book_type in self.borrowed_books:
                print(f"{self.name}, you have {self.borrowed_books[book_type]} {book_type} book(s) with you.")
            else:
                print(f"{self.name}, you have 0 {book_type} book(s) with you.")
        else:
            if self.borrowed_books:
                print(f"{self.name}, you have {sum(self.borrowed_books.values())} book(s) with you.")
                for book_type, quantity in self.borrowed_books.items():
                    print(f"Books on {book_type}: {quantity}")
            else:
                print(f"{self.name}, you have 0 books with you.")

L1 = Library('Dhaka', {'Arts': 15, 'Fiction': 135, 'Politics': 2, 'Science': 11, 'Poetry': 15})
L1.details()
print("1----------------------")
r1 = Reader('Aladdin')
r1.borrow(L1, 'Arts', 'Fiction', 'Fiction', 'Politics')
print("2----------------------")
r1.borrow(L1, 'Politics', 'Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2 = Reader('Jasmine')
r2.borrow(L1, 'Politics', 'Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```

```python
#Task 8
class Cylinder:
    radius = 5
    height = 18
    def __init__(self, radius, height):
        print(f"Default radius = {Cylinder.radius} and height = {Cylinder.height}.")

        Cylinder.radius = radius
        Cylinder.height = height

        print(f"Updated: radius = {Cylinder.radius} and height = {Cylinder.height}.")

    @staticmethod
    def area(radius, height):
        area =  (2 * 3.1416 * (radius ** 2)) + (2 * 3.1416 * radius * height)
        print(f"Area: {area}")

    @staticmethod
    def volume(radius, height):
        volume = 3.1416 * (radius ** 2) * height
        print(f"Volume: {volume}")

    @classmethod
    def swap(cls, height, radius):
        return cls(radius, height)
    
    @classmethod
    def changeFormat(cls, str):
        radius, height = str.split("-")
        return cls(float(radius), float(height))


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 6
class Shape:
  def __init__(self, name='Default', height=0, base=0): 
    self.area = 0 
    self.name = name 
    self.height = height 
    self.base = base 
 
  def get_height_base(self): 
    return "Height: "+str(self.height)+", Base: "+str(self.base)

class triangle(Shape):
  def calcArea(self):
    self.area = (self.height * self.base) / 2

  def printDetail(self):
    return (f"Shape name: {self.name}" + 
            f"\n{super().get_height_base()}" +
            f"\nArea: {self.area}"
            )

class trapezoid(Shape):
  def __init__(self, name, height, base, side_a):
    super().__init__(name, height, base)
    self.side_a = side_a

  def calcArea(self):
    self.area = ((self.height + self.base) * self.area) / 2

  def printDetail(self):
    return (f"Shape name: {self.name}" + 
            f"\n{super().get_height_base()}, Side_A: {self.side_a}" +
            f"\nArea: {self.area}"
            )

tri_default = triangle() 
tri_default.calcArea() 
print(tri_default.printDetail()) 
print('--------------------------') 
tri = triangle('Triangle', 10, 5) 
tri.calcArea() 
print(tri.printDetail()) 
print('---------------------------') 
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea() 
print(trap.printDetail())
```

```python
#Task 4
import math
class TwoDVector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def add2DVectors(self, *vectors):
        for i in vectors:
            self.x += i.x
            self.y += i.y
    def print2DVector(self):
        if self.y >= 0:
            y = "+ "+str(self.y)
        else:
            y = str(self.y)
        print(f"{self.x}i {y}j")
        
class ThreeDVector(TwoDVector):
    def __init__(self, x, y, z):
        super().__init__(x, y)
        self.z = z

    def add3DVectors(self, *vectors):
        super().add2DVectors(*vectors)
        for i in vectors:
            self.z += i.z
    
    def print3DVector(self):
        if self.y >= 0:
            y = "+ "+str(self.y)
        else:
            y = str(self.y)

        if self.z >= 0:
                z = "+ "+str(self.z)
        else:
            z = str(self.z)

        print(f"{self.x}i {y}j {z}k")

    def calculateLength(self):
        length = math.sqrt((self.x ** 2) + (self.y ** 2) + (self.z ** 2))
        return length

    def multiplyScalar(self, m):
        self.x *= m
        self.y *= m
        self.z *= m

TwoDV1 = TwoDVector(5, 6)
TwoDV2 = TwoDVector(3, 7)
TwoDV3 = TwoDVector(1, 8)
print("===============")
TwoDV1.add2DVectors(TwoDV2, TwoDV3)
TwoDV1.print2DVector()
print("===============")
ThreeDV1 = ThreeDVector(5, 6, 1)
ThreeDV2 = ThreeDVector(1, 9, -7)
ThreeDV3 = ThreeDVector(8, 2, 4)
print("===============")
ThreeDV1.add3DVectors(ThreeDV2,ThreeDV3)
ThreeDV1.print3DVector()
print("===============")
ThreeDV1.multiplyScalar(3)
ThreeDV1.print3DVector()
print("===============")
print(ThreeDV1.calculateLength())
```

```python
#Task 2
class RealNumber: 
    def __init__(self, number=0): 
        self.number = number 
    def __add__(self, anotherRealNumber): 
        return self.number + anotherRealNumber.number 
    def __sub__(self, anotherRealNumber): 
        return self.number - anotherRealNumber.number 
    def __str__(self): 
        return str(self.number) 
     
class ComplexNumber(RealNumber):
    def __init__(self, real, i):

        if isinstance(real, int):
            self.r = super().__init__(real)
            self.i = i
        else:
            self.r = super().__init__(real.number)
            self.i = i

    def __add__(self, other):
        real_part = self.number + other.number
        image_part = self.i + other.i
        return ComplexNumber(real_part, image_part)
    
    def __sub__(self, other): 
        real_part = self.number - other.number
        image_part = self.i - other.i
        return ComplexNumber(real_part, image_part)
    
    def __str__(self):
        if self.i < 0:
            return f"{(self.number)} - {-(self.i)}i"
        return f"{self.number} + {self.i}i"

r1 = RealNumber(3)
r2 = RealNumber(5) 
print(r1+r2) 
cn1 = ComplexNumber(2, 1) 
print(cn1) 
cn2 = ComplexNumber(r1, 5) 
print(cn2) 
cn3 = cn1 + cn2 
print(cn3) 
cn4 = cn1 - cn2 
print(cn4) 
```

```python
#Task 6
class Shape3D:
  pi = 3.14159 
  def __init__(self, name = 'Default', radius = 0): 
    self._area = 0 
    self._name = name 
    self._height = 'No need' 
    self._radius = radius    
 
  def calc_surface_area(self): 
    return 2 * Shape3D.pi * self._radius 
 
  def __str__(self): 
    return "Radius: "+str(self._radius)

class Sphere(Shape3D):
    def __init__(self, name, radius):
        super().__init__(name, radius)
        print(f"Shape name: {self._name}, Area Formula: 4 * pi * r * r")

    def calc_surface_area(self):
      self._area = super().calc_surface_area() * 2 * self._radius
      return self._area

    def __str__(self):
      return f"{super().__str__()}, Height: {self._height} \nArea: {self._area}"
   
class Cylinder(Shape3D):
    def __init__(self, name, radius, height):
        super().__init__(name, radius)
        self._height = height
        print(f"Shape name: {self._name}, Area Formula: 2 * pi * r * (r + h)")

    def calc_surface_area(self):
      self._area = super().calc_surface_area() * (self._radius + self._height)
      return self._area
   
    def __str__(self):
      return f"{super().__str__()}, Height: {self._height} \nArea: {self._area}"
 
 
sph = Sphere('Sphere', 5) 
print('----------------------------------') 
sph.calc_surface_area() 
print(sph) 
print('==================================') 
cyl = Cylinder('Cylinder', 5, 10) 
print('----------------------------------') 
cyl.calc_surface_area() 
print(cyl)
```

```python
#Task 3
class Account:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
        self.account_type = "General"
        self.maturity = 0

    def print_details(self):
        print("------ Account details ------")
        print(f"Account Type: {self.account_type}, Maturity: {self.maturity} years")
        print(f"Account Number: {self.account_number}, Balance: ${self.balance:.2f}")

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount:.2f}. New Balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
        else:
            print("Insufficient funds.")

    def year_passed(self, year):
        self.maturity += year
        print(f"Maturity of the account: {self.maturity} years")

class SavingsAccount(Account):
    def __init__(self, a_type, account_number, balance, int_rate, min_limit):
        super().__init__(account_number, balance)
        self.account_type = a_type
        self.int_rate = int_rate
        self.min_limit = min_limit

    def print_details(self):
        super().print_details()
        print(f"Interest Rate: {self.int_rate}, Minimum Limit: ${self.min_limit}")

    def apply_interest(self):
        self.balance += (self.balance * self.int_rate)
        print(f"Interest applied. New Balance: ${self.balance}")
    
    def withdraw(self, amount):
        if self.balance - amount < self.min_limit:
            print("Insufficient funds.")
        else:
            super().withdraw(amount)     


class FixedDepositAccount(Account):
    def __init__(self, a_type, account_number, balance, mat_age):
        super().__init__(account_number, balance)
        self.account_type = a_type
        self.mat_age = mat_age

    def deposit(self, amount):
        print("You can not deposit in a fixed deposit account.")

    def withdraw(self, amount):
        if self.maturity < self.mat_age:
            print("Can not withdraw, Account is not matured")
        else:
            if self.balance >= amount:
                self.balance -= amount
                print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
            else:
                print("Insufficient funds.")

print("-----------1------------")
account = Account("A203", 2000)
account.print_details()
print("-----------2------------")
account.deposit(400)
account.withdraw(1500)
account.year_passed(2)
print("-----------3------------")
account.print_details()
print("-----------4------------")
savings_account = SavingsAccount("Savings","SA123", 1000, 0.05, 500)
savings_account.print_details()
print("-----------5------------")
savings_account.deposit(400)
print("-----------6------------")
savings_account.withdraw(1000)
print("-----------7------------")
savings_account.withdraw(800)
print("-----------8------------")
savings_account.apply_interest()
print("-----------9------------")
savings_account.print_details()
print("-----------10------------")
fixed_account1= FixedDepositAccount("Fixed Deposit","FDA321", 10000, 5)
fixed_account1.print_details()
print("-----------11------------")
fixed_account1.deposit(400)
print("-----------12------------")
fixed_account1.year_passed(6)
print("-----------13------------")
fixed_account1.withdraw(10000)
print("-----------14------------")
fixed_account1.print_details()
print("-----------15------------")
fixed_account2 = FixedDepositAccount("Fixed Deposit","FDA300", 50000, 7)
fixed_account2.print_details()
print("-----------16------------")
fixed_account2.withdraw(10000)
```

```python
#Q3
class Workspace:
    def __init__(self, workspaceName, type):
        self.workspaceName = workspaceName
        self.type = type
        self.channels = {}

    def checkChannelExistence(self, channelName):
        if channelName in self.channels:
            return True
        return False
    
    def __str__(self):
        return f"Name: {self.workspaceName}\nType:{self.type}\n"
    
class SlackWorkspace(Workspace):
    total_workspaces = 0
    def __init__(self, workspaceName, type, c_limit = 3):
        super().__init__(workspaceName, type)
        self.members = 0
        self.mem_count = 0
        self.c_limit = c_limit
        self.c_count = 0
        print(f"{self.workspaceName} is created")

        SlackWorkspace.total_workspaces += 1
    
    def createChannels(self):
        for i  in range(1, self.c_limit+1):
            self.channels[f"Section-{i}"] = []
            print(f"Section-{i} channel created")
            self.c_count += 1

    def addMembersToChannel(self, *args):
        for i in args:
            name, sec = i.split('_')
            sec = f"Section-{sec}"
            if super().checkChannelExistence(sec):
                self.channels[sec].append(name)
                self.mem_count += 1
                print(f"{name} added to {sec} channel")
            else: print(f"Sorry {name}! {sec} channel not found in {self.workspaceName}")

    def __str__(self):
        return(
            f"{super().__str__()}"
            f"Total members: {self.members}\n"
            f"Total Channels: {self.c_count}\n"
            f"Channel info: {self.channels}"
)    
    
print(f"Total no. of workspaces: {SlackWorkspace.total_workspaces}")
print("1=============================")
CSE111Workspace = SlackWorkspace("CSE111Workspace", "Education",5)
print("2=============================")
print(CSE111Workspace)
print(f"Total no. of workspaces: {SlackWorkspace.total_workspaces}")
print("3=============================")
CSE111Workspace.createChannels()
print("4=============================")
print(CSE111Workspace)
print("5=============================")
john, harry, adil = "John_2", "Harry_10", "Adil_1"
alice, bob = "Alice_2","Bob_3"
print("6=============================")
CSE111Workspace.addMembersToChannel(john, harry)
CSE111Workspace.addMembersToChannel(adil)
print("7=============================")
del(CSE111Workspace.channels["Section-3"])
print(CSE111Workspace)
print("8=============================")
CSE111Workspace.addMembersToChannel(alice, bob)
print("9=============================")
print(CSE111Workspace)
print("10=============================")
CSE320Workspace = SlackWorkspace("CSE320Workspace",
"Education")
CSE320Workspace.createChannels()
print("11=============================")
print(CSE320Workspace)
print("12=============================")
print(f"Total no. of workspaces: {SlackWorkspace.total_workspaces}")
```
