```python
#Task 1

list1 = []

while True:
  user = input()
  if user == "STOP":
    break
  else:
    list1.append(int(user))

print()

for i in range(len(list1)):
  if list1[i] not in list1[:i]:
    counter = 1
    for j in range(i+1,len(list1)):
      if list1[i] == list1[j]:
        counter += 1
    print(list1[i],"-",counter,"times")
```

```python
#Task 1 (Class)
uniq_list = []
freq_list = []
all_list = []

while True:
    user = input()
    if user == "STOP":
        break
    else:
        all_list.append(int(user))
        if int(user) not in uniq_list:
            uniq_list.append(int(user))

for i in uniq_list:
    counter = 0
    for j in all_list:
        if j == i:
            counter += 1
    freq_list.append(counter)

for k in range(len(freq_list)):
    print(uniq_list[k],"-",freq_list[k],"times")
```

```python
#Task 2
N = int(input("Type the value of N: "))
lists = []
sum = 0
sum_list = []

for i in range(N):
    user_input = input("Enter the list values separated by space: ")
    current_list = [int(val) for val in user_input.split()]
    lists.append(current_list)

print(lists)
print()

for j in range(len(lists)):
    temp_sum = 0
    for k in range(len(lists[j])):
        temp_sum += int(lists[j][k])
        if temp_sum > sum:
            sum = temp_sum
            sum_list = lists[j]

print("highest sum :",sum)
print()
print("list with the highest sum :",sum_list)
```

```python
#Task 5
input_str = input("type any word")
chars = list(input_str)
new_str = []

# Process lowercase letters
for char in sorted(chars):
    if char.islower():
        new_str.append(char)

# Process uppercase letters
for char in sorted(chars):
    if char.isupper():
        new_str.append(char)

# Process odd digits
for char in sorted(chars):
    if char.isdigit() and int(char) % 2 != 0:
        new_str.append(char)

# Process even digits
for char in sorted(chars):
    if char.isdigit() and int(char) % 2 == 0:
        new_str.append(char)

# Concatenate the results
result = ''.join(new_str)
print(result)
```

```python
#Task 6
n = int(input())  # Number of students
k = int(input())  # Minimum number of times each team should participate in the world championship
lst = []  # List to store the participation history of each student

# Input the participation history for each student
for i in range(n):
    np = int(input())
    lst.append(np)

student = 0  # Variable to count the eligible students

# Count the number of students who can participate in a team
for i in lst:
    if i + k <= 5:
        student += 1

team = student // 3  # Calculate the maximum number of teams that can be formed

print(team)
```

```python
#Task 1(Class)
d1 = {'a':100,'b':100,'c':200,'d':300}
d2 = {'a':300,'b':200,'d':400,'e':200}

for k, v in d2.items():
  if k not in d1.keys():
    d1[k] = v
  else:
    d1[k] = d1[k] + v
print(d1)

tuple1 = []
for k in d1.values():
  tuple1 += [k]
tuple1 = tuple(sorted(tuple1))
print(tuple1)
```

```python
#Task 5

dict1 = {}
dict2 = {}
inp1 = input("Enter the first string: ")
inp2 = input("Enter the second string: ")

for i in inp1:
  if i not in dict1:
    dict1[i] = 1
  else:
    dict1[i]+=1
for j in inp2:
  if j not in dict2:
    dict2[j] = 1
  else:
    dict2[j]+=1

if dict1==dict2:
  print("Those strings are anagrams")
else:
  print("Those strings are not anagrams")
```
