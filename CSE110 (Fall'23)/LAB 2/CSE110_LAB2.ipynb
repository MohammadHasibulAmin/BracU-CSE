```python
#Class_Evaluation_Task_1

x = int(input("hours worked : "))

if x>=40 and x <=168:
  print("salary : ", str(8000+(x-40)*300)+" for "+str(x)+" hours worked")
elif x<40 and x>=0:
  print(x*200)
elif x<0:
  print("work hours cannot be negative")
else:
  print("Impossible to work more than 168 hours weekly")
```
