```python
#Class_Task_5

x = int(input("Please enter number of rows: "))
y = int(input("Please enter number of columns : "))

current_row = 1
while current_row <= x:
    current_col = 1
    while current_col <= y:
        print(current_col, end="")
        current_col += 1
    print()
    current_row += 1
```

```python
#Home_Task_3

num = int(input("enter any number : "))

count = 0
sum = 0
for i in range(2,num):
    if num%i == 0:
        count+=1
if count==0:
    print(num,"is a prime number")
else:
    for i in range(1,num):
        if num%i==0:
            sum+=i
    if sum==num:
        print(num,"is a perfect number")
    else:
        print(num,"is not a prime or a perfect number")
```
